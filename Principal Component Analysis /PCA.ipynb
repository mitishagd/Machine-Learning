```python
import scipy.io as sio
import numpy as np
from bokeh.plotting import figure, show, output_notebook
import matplotlib.pyplot as plt
```

```python
content = sio.loadmat('dataset3.mat')
```

```python
# The given Y matrix is 40*200. Transpose it to perform PCA.
Y = np.transpose(np.array(content['Y']))
```

```python
Y.shape
```

```python
def PCA(X, k):
    (n, m) = X.shape
    sigma = 1.0 / m * np.dot(X ,np.transpose(X))
    U, s, V = np.linalg.svd(sigma, full_matrices = True)
    U_reduced = U[:, : k]
    Z = np.dot(np.transpose(U_reduced) ,X)
    return np.transpose(Z), U_reduced
```

```python
projection, components = PCA(Y, 2)
```

```python
components.shape
```

```python
x_component = []
y_component = []
for component in components:
    x_component.append(component[0])
    y_component.append(component[1])
```

```python
plt.plot(x_component, y_component, 'ro')
plt.show()
```

From the graph above, we can observe that the graph is fairly scattered.
